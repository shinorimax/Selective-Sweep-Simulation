```python
import tskit, pyslim

# Load tree sequence
ts = tskit.load("trees/r3020002.trees")

# Decode metadata using the schema built into pyslim
mut_schema = pyslim.metadata_mutation_schema  # this knows how to interpret SLiM metadata

sel_positions = []

for mut in ts.tables.mutations:
    md = mut_schema.decode_row(mut.metadata)
    if md['selection_coeff'] != 0 or md['mutation_type'] == 3:
        pos = ts.site(mut.site).position
        sel_positions.append(pos)

print(sorted(set(sel_positions)))
```
